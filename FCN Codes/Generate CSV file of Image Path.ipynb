```python
def prepare_csv(image_dir_path, labeled_image_dir_path, saving_dir_path, csv_filename, type):
  import os
  images = os.listdir(image_dir_path)
  images_path = []
  labeled_images_path = []
  for image in images:
    images_path.append(image_dir_path + "/" + image)
    labeled_images_path.append(labeled_image_dir_path +"/" + image.split(".")[0] + "_L." + type)
  import pandas as pd
  df = pd.DataFrame({'Image Paths':images_path, 'Labeled Image Paths':labeled_images_path})
  df.to_csv(saving_dir_path + "/" + csv_filename + ".csv", index = False)
```

# Prepare Training CSV

```python
train_image_dir_path = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/CamVid/train"
train_labeled_image_dir_path = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/CamVid/train_labels_npy"
saving_dir_path = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/CamVid/csv_files"
train_csv_filename = "train"
prepare_csv(train_image_dir_path, train_labeled_image_dir_path, saving_dir_path, train_csv_filename, "npy")
```

# Prepare Validation CSV

```python
val_image_dir_path = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/CamVid/val"
val_labeled_image_dir_path = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/CamVid/val_labels_npy"
saving_dir_path = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/CamVid/csv_files"
val_csv_filename = "val"
prepare_csv(val_image_dir_path, val_labeled_image_dir_path, saving_dir_path, val_csv_filename, "npy")
```

# Prepare Validation CSV

```python
test_image_dir_path = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/CamVid/test"
test_labeled_image_dir_path = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/CamVid/test_labels_npy"
saving_dir_path = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/CamVid/csv_files"
test_csv_filename = "test"
prepare_csv(test_image_dir_path, test_labeled_image_dir_path, saving_dir_path, test_csv_filename, "npy")
```
