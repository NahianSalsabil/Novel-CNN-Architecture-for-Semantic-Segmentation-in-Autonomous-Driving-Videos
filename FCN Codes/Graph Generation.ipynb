```python
wavelet_metrics_file = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/metrics/wavelet_metrics.txt"
with open(wavelet_metrics_file, "r") as file:
  content = file.read()
  print(content)
```

```python
w_pixel_accuracy = []
w_mIoU = []
w_training_loss = []
epoch = []
with open(wavelet_metrics_file, "r") as file:
  for line in file:
    l = line
    epoch.append(float(l.split("..  ")[0].split(": ")[1].split("/")[0]))
    w_pixel_accuracy.append(float(l.split("..  ")[-1].split(".. ")[1].split(" ")[0]))
    w_mIoU.append(float(l.split(".. ")[-1].split("\n")[0]))
    w_training_loss.append(float(l.split("..  ")[1].split(": ")[1]))
print(epoch)
print(w_pixel_accuracy)
print(w_mIoU)
print(w_training_loss)
```

```python
wavelet_metrics_file = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/metrics/fcn_metrics.txt"
with open(wavelet_metrics_file, "r") as file:
  content = file.read()
  print(content)
```

```python
fcn_pixel_accuracy = []
fcn_mIoU = []
fcn_training_loss = []
epoch = []
with open(wavelet_metrics_file, "r") as file:
  for line in file:
    l = line
    epoch.append(float(l.split("..  ")[0].split(": ")[1].split("/")[0]))
    fcn_pixel_accuracy.append(float(l.split("..  ")[-1].split(".. ")[1].split(" ")[0]))
    fcn_mIoU.append(float(l.split(".. ")[-1].split("\n")[0]))
    fcn_training_loss.append(float(l.split("..  ")[1].split(": ")[1]))
print(epoch)
print(fcn_pixel_accuracy)
print(fcn_mIoU)
print(fcn_training_loss)
```

```python
wavelet_metrics_file = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/metrics/pixelbinning_metrics_v2.txt"
with open(wavelet_metrics_file, "r") as file:
  content = file.read()
  print(content)
```

```python
p_pixel_accuracy = []
p_mIoU = []
p_training_loss = []
epoch = []
with open(wavelet_metrics_file, "r") as file:
  for line in file:
    l = line
    epoch.append(float(l.split("..  ")[0].split(": ")[1].split("/")[0]))
    p_pixel_accuracy.append(float(l.split("..  ")[-1].split(".. ")[1].split(" ")[0]))
    p_mIoU.append(float(l.split(".. ")[-1].split("\n")[0]))
    p_training_loss.append(float(l.split("..  ")[1].split(": ")[1]))
print(epoch)
print(p_pixel_accuracy)
print(p_mIoU)
print(p_training_loss)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (5, 4))
plt.plot(epoch, w_pixel_accuracy, color = 'r', label = 'Wavelet Pooling', linewidth=1)
plt.plot(epoch, fcn_pixel_accuracy, color = 'b', label = 'FCN-Max Pooling', linewidth=1)
plt.xlabel("No. of Epoch")
plt.ylabel("Pixel Accuracy")
plt.title("")
plt.legend()

plt.savefig("/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/metrics/graphs/fcn_vs_wavelet_pacc.jpeg", bbox_inches = 'tight', dpi=1200)
plt.show()
```

```python
plt.figure(figsize = (5, 4))
plt.plot(epoch, w_mIoU, color = 'r', label = 'Wavelet Pooling', linewidth=1)
plt.plot(epoch, fcn_mIoU, color = 'b', label = 'FCN-Max Pooling', linewidth=1)
plt.xlabel("No. of Epoch")
plt.ylabel("Mean IoU")
plt.title("")
plt.legend()

plt.savefig("/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/metrics/graphs/fcn_vs_wavelet_iou.jpeg", bbox_inches = 'tight', dpi=1200)
plt.show()
```

```python
plt.figure(figsize = (5, 4))
plt.plot(epoch, w_training_loss, color = 'r', label = 'Wavelet Pooling', linewidth=1)
plt.plot(epoch, fcn_training_loss, color = 'b', label = 'FCN-Max Pooling', linewidth=1)
plt.xlabel("No. of Epoch")
plt.ylabel("Training Loss")
plt.title("")
plt.legend()

plt.savefig("/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/metrics/graphs/fcn_vs_wavelet_loss.jpeg", bbox_inches = 'tight', dpi=1200)
plt.show()
```
