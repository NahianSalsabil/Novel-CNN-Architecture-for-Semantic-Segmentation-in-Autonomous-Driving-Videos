**Prepare Train CSV**

```python
import os
from glob import glob
train_dir = "/content/drive/My Drive/Thesis (Image Segmentation)/FCN/CamVid/train"
train_labeled_dir = "/content/drive/My Drive/Thesis (Image Segmentation)/FCN/CamVid/train_labels"
```

```python
type = "*.png"
train_images = os.listdir(train_dir)
train_labeled_images = []
print(len(train_images))
for image in train_images:
  train_labeled_images.append(image.split(".")[0] + "_L." + image.split(".")[1])
print(len(train_labeled_images))
print(train_labeled_images[0])
```

```python
print(f' train images : {len(train_images)} items')
print(f' train labeled images : {len(train_labeled_images)} items')
print("train:")
print(train_images)
print("labeled:")
print(train_labeled_images)
```

```python
import pandas as pd
df = pd.DataFrame({'Train Images':train_images, 'Train Labeled Images':train_labeled_images})
df
```

```python
df.to_csv("/content/drive/My Drive/Thesis (Image Segmentation)/FCN/CamVid/CSV/" + "train.csv", index = False)
```
