# Create Function for both of the above sections

```python
def load_image_name_file(image_directory_path):
  from PIL import Image
  import glob
  image_list = []
  for filename in glob.glob(image_directory_path+'/*.png'):
      img=Image.open(filename)
      image_list.append((filename.split('/')[-1].split('.')[0], img))
  return image_list
```

```python
def convert_to_npy(image_list, npy_directory_path):
  import numpy as np
  for img in image_list:
    img_array = np.array(img[1])
    label_seg = np.zeros((img_array.shape[:2]), dtype=int)
    for k in color2index:
      label_seg[(img_array==k).all(axis=2)] = color2index[k]
    np.save(npy_directory_path +'/' + img[0], label_seg)
```

# Using Above Functions

```python
train_labels_path = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/CamVid/train_labels"
test_labels_path = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/CamVid/test_labels"
val_labels_path = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/CamVid/val_labels"
train_labels_npy_path = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/CamVid/train_labels_npy"
test_labels_npy_path = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/CamVid/test_labels_npy"
val_labels_npy_path = "/content/drive/MyDrive/Thesis (Image Segmentation)/FCN/CamVid/val_labels_npy"
convert_to_npy(load_image_name_file(train_labels_path), train_labels_npy_path)
convert_to_npy(load_image_name_file(test_labels_path), test_labels_npy_path)
convert_to_npy(load_image_name_file(val_labels_path), val_labels_npy_path)
```
